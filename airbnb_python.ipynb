```python


# 1. Handle categorical missing values
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')

# 2. Handle numerical missing values
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# 3. Convert last_review to datetime (explicit format to avoid warnings)
df['last_review'] = pd.to_datetime(
    df['last_review'],
    format='%d-%m-%Y',
    errors='coerce'
)

# 4. Create review existence indicator
df['has_review'] = df['last_review'].notnull().astype(int)
```
